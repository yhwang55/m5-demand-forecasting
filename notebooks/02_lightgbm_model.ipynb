# 02 LightGBM Model

Feature-based model for tabular time series


```python
from src.pipeline import build_sample_dataset
from src.models.lightgbm_model import LightGBMModel
from src.metrics import rmse, mae, mape

df = build_sample_dataset()
X = df[["year","month","week","day","dow"]]
y = df["sales"]
model = LightGBMModel()
model.fit(X, y)
preds = model.predict(X)
rmse(y, preds), mae(y, preds), mape(y, preds)
```
